# **About Dataset**


**Description**

Survey data on smoking habits from the UK. The data set can be used for analyzing the demographic characteristics of smokers and types of tobacco consumed.

**Format**

A data frame with 1691 observations on the following 12 variables.

- gender
Gender with levels Female and Male.

- age
Age.

- marital_status
Marital status with levels Divorced, Married, Separated, Single and Widowed.

- highest_qualification
Highest education level with levels A Levels, Degree, GCSE/CSE, GCSE/O Level, Higher/Sub Degree, No Qualification, ONC/BTEC and Other/Sub Degree

- nationality
Nationality with levels British, English, Irish, Scottish, Welsh, Other, Refused and Unknown.

- ethnicity
Ethnicity with levels Asian, Black, Chinese, Mixed, White and Refused Unknown.

- gross_income
Gross income with levels Under 2,600, 2,600 to 5,200, 5,200 to 10,400, 10,400 to 15,600, 15,600 to 20,800, 20,800 to 28,600, 28,600 to 36,400, Above 36,400, Refused and Unknown.

- region
Region with levels London, Midlands & East Anglia, Scotland, South East, South West, The North and Wales

- smoke
Smoking status with levels No and Yes

- amt_weekends
Number of cigarettes smoked per day on weekends.

- amt_weekdays
Number of cigarettes smoked per day on weekdays.

- type
Type of cigarettes smoked with levels Packets, Hand-Rolled, Both/Mainly Packets and Both/Mainly Hand-Rolled

- Source
National STEM Centre, Large Datasets from stats4schools, https://www.stem.org.uk/resources/elibrary/resource/28452/large-datasets-stats4schools.


Link to Dataset: https://www.kaggle.com/datasets/utkarshx27/smoking-dataset-from-uk

```python
df = pd.read_csv('D:\\zerotopandas\\smoking.csv')
```

# Data Cleaning

```python
df
```

```python
# Drop colum
df.drop(columns = 'Unnamed: 0',inplace = True)
```

```python
# Check Data types
df.info()
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
# Drop duplicate values
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
# Check null values
print('Number of null values')
print('---------------------')
print(df.isnull().sum())
print('Percentage of null values')
print('---------------------')
print(df.isnull().mean()*100)
```

```python
# Impute missing values
df['amt_weekends']= np.where(df['smoke'] == 'No' , df['amt_weekends'] == 0 ,  df['amt_weekends'])
df['amt_weekdays']= np.where(df['smoke'] == 'No' , df['amt_weekdays'] == 0 ,  df['amt_weekdays'])
```

```python
#Check unique values
for i in df.columns:
    print('Column "{}" has the following unique values: {}'.format(i,df[i].unique()))
    print('and the number of unique values is/are : {}'.format(df[i].nunique()))
    print('*********************************************************************')
```

```python
# Transform gross_income columns
gross = df['gross_income'].str.split(' ',expand = True)
gross[0] = gross[0].str.replace(',','').replace('Under',0).replace('Above',0).replace('Refused',0).replace('Unknown',0)
gross[1] = gross[1].str.replace(',','').replace('to',0)
gross[2] = gross[2].str.replace(',','').replace('None',0)

# Change dtypes
gross[0] = pd.to_numeric(gross[0],errors = 'coerce')
gross[1] = pd.to_numeric(gross[1],errors = 'coerce')
gross[2] = pd.to_numeric(gross[2],errors = 'coerce')
# Fill Na
gross = gross.fillna(0)
# Calculate average income
gross[3] = (gross[0] +gross[2])/2
# Impute 0 with constant number from column 1
gross[3] = np.where(gross[3] == 0 , gross[1],gross[3] )
# Drop unused columns
gross = gross.drop(columns = [0,1,2])
# Concat colculated column with df
df = pd.concat([df,gross] , axis = 1)
#
# Rename column
df = df.rename(columns = {3 :'average_income'})
# Drop the gross_income column
df = df.drop(columns = 'gross_income')
```
